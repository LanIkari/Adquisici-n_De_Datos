```python
#Cambiaremos los Nulos 
df['calificacion'].fillna(value=0, inplace=True)
df[df.index.isin(df_na_calif.index)]

#Con promedio 
##df['calificacion'].fillna(value=df['calificacion'].mean(), inplace=True)
df['calificacion'].fillna(value=df['calificacion'].min(), inplace=True)
df[df.index.isin(df_na_calif.index)]
```

```python
print ('Promedio', df['calificacion'].mean())
print ('Max', df['calificacion'].max())
print ('Min', df['calificacion'].min())
print ('No Calificaciones', df['calificacion'].count())
print ('Sum total', df['calificacion'].sum())
```

```python
#Generamos nuevas columnas, en este caso _gpa_. Si ya exste lo cambia, si no existe lo genera
df['gpa']= (((df['calificacion']*100)/df['calificacion'].max())/20)-1
df.sort_values(by='calificacion', ascending=False)
```
